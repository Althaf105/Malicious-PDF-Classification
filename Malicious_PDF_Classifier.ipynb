# Parse and Annotate All Pdf files

```python
benign_dir = "benign_files\Benign\Ben01"
malicious_dir = "mal_pdfs\Malicious\\f1\\f1"

labelled_dataset = []

try:
    #annotate benign pdfs
    for file in os.listdir(benign_dir):
        command_to_execute = 'python pdfid.py ' + 'benign_files\Benign\Ben01/' + '"' + file + '"'
        try:
            stdout = Popen(command_to_execute, shell=True, stdout=PIPE).stdout
            output = stdout.readlines()
            labelled_dataset.append(getFeatures(output, 'no'))
        except Exception as e:
            print("Issue with file:",file)
            continue
    #annotate malicious pdfs
    for file in os.listdir(malicious_dir):
        command_to_execute = 'python pdfid.py ' + 'mal_pdfs\Malicious\f1\f1/' + '"' + file + '"'
        try:
            stdout = Popen(command_to_execute, shell=True, stdout=PIPE).stdout
            output = stdout.readlines()
            labelled_dataset.append(getFeatures(output, 'yes'))
        except Exception as e:
            print("Issue with file:",file)
            continue
except Exception as ex:
    print("File failed", file)
    pass
```

# save dataset to csv file

```python
import pandas as pd
featurename_list = ['obj', 'stream', 'xref', 'startxref', 'Encrypt', 'ObjStm', 'JS', 'Javascript', 'AA', 'OpenAction', 'AcroForm', 'JBIG2Decode', 'RichMedia',
            'Launch', 'EmbeddedFile', 'XFA', 'Malicious_Label']
labelled_df = pd.DataFrame(labelled_dataset,columns=featurename_list)
labelled_df_shuffled = labelled_df.sample(frac=1).reset_index(drop=True)
labelled_df_shuffled.to_csv('pdf_labelleddataset.csv',index=False)
```

# Normalize data

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
features = featurename_list[:-1]
scalar = StandardScaler()
# the fit_transform ops returns a 2d numpy.array, we cast it to a pd.DataFrame
standardized_features = pd.DataFrame(scalar.fit_transform(labelled_df_shuffled[features].copy()), columns = features)
old_shape = labelled_df_shuffled.shape
# drop the unnormalized features from the dataframe
labelled_df_shuffled.drop(features, axis = 1, inplace = True)
# join back the normalized features
df = pd.concat([labelled_df_shuffled, standardized_features], axis= 1)
assert old_shape == df.shape, "something went wrong!"
```

```python
df.to_csv('pdf_normalizeddataset.csv',index=False)
```

# Supervised Learning Classifier

```python
from pandas import read_csv
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
# from feature_extraction import feature_extraction

#using normalized data for training
df = pd.read_csv('pdf_normalizeddataset.csv')
X = df.iloc[:, 1: 17]
y = df.iloc[:, 0]

#using unnormalized data for training
# df = pd.read_csv('pdf_labelleddataset.csv')
# X = df.iloc[:, 0:16]
# y = df.iloc[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
print("---Random Forest---")
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
acs = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
print("Accuracy:", acs)
print("\nConfusion Matrix:\n", cm)
```
